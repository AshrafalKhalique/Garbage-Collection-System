```python
class User:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def receive_warning_message(self, message):
        print("Warning:", message)

    def check_total_bill(self, bill):
        print("Total Bill:", bill, "tk.")

class BillingInfo:
    def __init__(self):
        self.biodegradable_item_cost = 0
        self.recyclable_item_cost = 2
        self.non_recyclable_item_cost = 5

class Bin:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = 0
        self.status = "Empty"

    def receive_garbage(self, amount):
        self.items += amount
        if self.items >= self.capacity:
            self.status = "Full"

    def send_warning_message(self, message):
        print("Warning:", message)

class SourceBin(Bin):
    def __init__(self, capacity):
        super().__init__(capacity)

    def update_status(self):
        if self.items > 0:
            self.status = "Partially filled"
        else:
            self.status = "Empty"

class GMPBin(Bin):
    def __init__(self, capacity):
        super().__init__(capacity)

    def update_status(self):
        if self.items >= self.capacity:
            self.status = "Full"
        else:
            self.status = "Partially filled"

class BBin:
    def __init__(self):
        self.items = 0

class NBBin(Bin):
    def __init__(self, capacity):
        super().__init__(capacity)

class RBin:
    def __init__(self):
        self.items = 0

class NRBin:
    def __init__(self):
        self.items = 0

# Sample data for demonstration
users = {
    "Ashraf": "123"
}

def login():
    username = input("Enter username: ")
    password = input("Enter password: ")
    if username in users and users[username] == password:
        return True
    else:
        print("Invalid username or password.")
        return False

def main():
    if login():
        name = input("Enter your name: ")
        address = input("Enter your address: ")
        user = User(name, address)

        billing_info = BillingInfo()
        source_bin = SourceBin(10)
        gmp_bin = GMPBin(10)
        b_bin = BBin()
        nb_bin = NBBin(10)
        r_bin = RBin()
        nr_bin = NRBin()

        biodegradable = int(input("Enter the amount of biodegradable garbage: "))
        recyclable = int(input("Enter the amount of recyclable garbage: "))
        non_recyclable = int(input("Enter the amount of non-recyclable garbage: "))

        # Simulating garbage collection process
        source_bin.receive_garbage(biodegradable + recyclable + non_recyclable)
        source_bin.update_status()
        if source_bin.status == "Full":
            user.receive_warning_message("Reduce garbage production")

        # Updating billing catalog
        b_bin.items += biodegradable
        r_bin.items += recyclable
        nr_bin.items += non_recyclable

        total_bill = (
            billing_info.biodegradable_item_cost * b_bin.items +
            billing_info.recyclable_item_cost * r_bin.items +
            billing_info.non_recyclable_item_cost * nr_bin.items
        )
        user.check_total_bill(total_bill)

if __name__ == "__main__":
    main()


# Â© BY ASHRAF AL- KHALIQUE
```
